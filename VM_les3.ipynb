```python
from scipy.optimize import fsolve

x = np.linspace(-50, 50, 1000)
plt.grid(True)
plt.ylim(-10, 40)
plt.xlim(-6, 6)
plt.plot(x, x**2-1)
plt.plot(x, (np.exp(x)+x-1)/x, color = 'green')
plt.show()
```

```python
def equations(p):
    x,y = p
    return(np.exp(x)+x-x*y-1, y-x**2+1)
x1,y1 = fsolve(equations,(-2,4))
print(x1,y1)
```

#### Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.


```python

def polar(R, a):
    x = R * np.cos(a)
    y = R * np.sin(a)
    return x, y
```

```python
t = np.arange(-2 * np.pi, 2 * np.pi, 0.01)
r = 1
x = r * np.cos(t)
y = r * np.sin(t)

plt.polar(x, y)
plt.show()
```
